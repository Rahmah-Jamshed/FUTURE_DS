Importing Libraries 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

Data Preprocessing 

```python
df = pd.read_csv("student_feedback.csv")
df.info()
```

```python
df.shape
```

```python
df.head()
```

Data Cleaning 

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Drop extra columns
df_clean = df.drop(columns=['Unnamed: 0', 'Student ID'])
df_clean.head()
```

```python
# Average rating for each question
avg_scores = df_clean.mean().sort_values(ascending=False)

print(avg_scores)
```

```python
# Bar chart of averages
plt.figure(figsize=(10,6))
avg_scores.plot(kind='bar', color="purple")
plt.title("Average Ratings per Feedback Category", fontsize=14)
plt.ylabel("Average Score (1-10)")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Data Visualization 

```python
#Distribution of Ratings
plt.figure(figsize=(12,8))
sns.boxplot(data=df_clean, palette="coolwarm")
plt.title("Distribution of Ratings for Each Question", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Score (1-10)")
plt.show()
```

```python
#Correlation Between Questions
plt.figure(figsize=(10,8))
sns.heatmap(df_clean.corr(), annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Between Feedback Categories")
plt.show()
```
